```python
# 读取ks_ds_label.csv
import pandas as pd


ks_ds_label_res = pd.read_csv('output/ks_ds_label.csv')
# 选取is_negative为False的行
ks_ds_pos_res = ks_ds_label_res[ks_ds_label_res['is_negative'] == False]
# 选取is_negative为True的行
ks_ds_neg_res = ks_ds_label_res[ks_ds_label_res['is_negative'] == True]

# 输出pos_res前200个title
ks_ds_pos_examples = ks_ds_pos_res['title'].head(200)
print("-"*100)
print("非负面内容：")
for example in ks_ds_pos_examples:
    print(example)

# 输出neg_res前200个title
ks_ds_neg_examples = ks_ds_neg_res['title'].head(200)
print("-"*100)
print("负面内容：")
for example in ks_ds_neg_examples:
    print(example)

# 输出neg_res整体比例
neg_ratio = ks_ds_neg_res['is_negative'].value_counts() / len(ks_ds_label_res)
print('\n负面比例：', neg_ratio)
print('负面数量：', len(ks_ds_neg_res))
```

```python
from api_neg_label import NEGATIVE_LABELS


# 遍历ks_ds_neg_res的每一行
neg_title_dict = {label: [] for label in NEGATIVE_LABELS}
for index, row in ks_ds_neg_res.iterrows():
    negative_labels = row['negative_labels'].split('|')
    for cur_label in negative_labels:
        neg_title_dict[cur_label].append(row['title'])

# 输出neg_title_dict
print('各主题负面标题数目：')
for label, titles in neg_title_dict.items():
    print(f'{label}：{len(titles)}')

# 输出样例
print('各主题负面样例：')
for label, titles in neg_title_dict.items():
    print(f'{label}：')
    for title in titles[:50]:
        print(title)
    print('-'*100)

# 输出负面样本到xlsx文件，用于人工打标确定precision
import pandas as pd
neg_rows = []
for label, titles in neg_title_dict.items():
    for title in titles:
        neg_rows.append({'title': title, 'label': label, 'human_judge': 1})
pd.DataFrame(neg_rows).to_excel('output/ks_ds_neg_title_dict.xlsx', index=False)

# 输出正面样本到xlsx文件，用于人工打标估计负面主题整体recall
pos_rows = []
for index, row in ks_ds_pos_res.head(1000).iterrows():
    pos_rows.append({'title': row['title'], 'human_judge': 0})
pd.DataFrame(pos_rows).to_excel('output/ks_ds_pos_title_dict.xlsx', index=False)
```

```python
# 通过人工打标结果计算负面内容整体召回率和各个负面主题的准确率

## 读取xlsx文件
ks_ds_pos_title_judged = pd.read_excel('output/ks_ds_pos_title_judged.xlsx')
ks_ds_neg_title_judged = pd.read_excel('output/ks_ds_neg_title_judged.xlsx')

## 估计全体负面内容召回率
ds_pos_num = len(ks_ds_pos_res)
false_pos_num = sum(ks_ds_pos_title_judged['human_judge'] == 1) * ds_pos_num / len(ks_ds_pos_title_judged)
true_neg_titles = ks_ds_neg_title_judged[ks_ds_neg_title_judged['human_judge'] == 1]['title']
true_neg_num = len(set(true_neg_titles))
overall_neg_recall = true_neg_num / (false_pos_num + true_neg_num)
print(f'全体负面内容召回率：{overall_neg_recall}')

## 各个负面主题的准确率
print('============各个负面主题的准确率============')
for label in NEGATIVE_LABELS:
    cur_neg_label_rows = ks_ds_neg_title_judged[ks_ds_neg_title_judged['label'] == label]
    cur_neg_label_num = len(cur_neg_label_rows)
    cur_neg_label_true_num = sum(cur_neg_label_rows['human_judge'] == 1)
    cur_neg_label_precision = cur_neg_label_true_num / cur_neg_label_num
    print(f'{label}准确率：{cur_neg_label_precision}')
```

```python
# 汇总人工打标结果，构建测试集
import pandas as pd
from collections import defaultdict


ks_pos_test_df = pd.read_excel('output/ks_ds_pos_title_judged.xlsx')
ks_pos_test_df = ks_pos_test_df[ks_pos_test_df['human_judge'] == 0]
ks_neg_test_df = pd.read_excel('output/ks_ds_neg_title_judged.xlsx')
ks_neg_test_df = ks_neg_test_df[ks_neg_test_df['human_judge'] == 1]

ks_hard_pos_test_df = ks_neg_test_df[ks_neg_test_df['human_judge'] == 0]
ks_pos_titles = ks_hard_pos_test_df['title'].tolist() + ks_pos_test_df['title'].tolist()

ks_hard_neg_test_df = ks_pos_test_df[ks_pos_test_df['human_judge'] == 1]
neg_title_label_dict = defaultdict(list)
for index, row in ks_hard_neg_test_df.iterrows():
    neg_title_label_dict[row['title']].append(row['label'])
for index, row in ks_neg_test_df.iterrows():
    neg_title_label_dict[row['title']].append(row['label'])

# 将人工打标的样本保存到xlsx文件
test_samples = []
for title, labels in neg_title_label_dict.items():
    negative_labels = '|'.join(labels)
    test_samples.append({'title': title, 'negative_labels': negative_labels, 'is_negative': 1})

for title in ks_pos_titles:
    test_samples.append({'title': title, 'negative_labels': '', 'is_negative': 0})

ks_test_df = pd.DataFrame(test_samples)
ks_test_df.to_excel('output/ks_ds_test.xlsx', index=False)
```

```python
import pandas as pd
from api_neg_label import NEGATIVE_LABELS

# 读取打标结果excel
res_path = 'output/ks_ds_v2_test_label_converted.xlsx'
ks_test_label_res = pd.read_excel(res_path)
ks_test_gt = pd.read_excel('output/ks_ds_test.xlsx')

# 计算负例整体召回
neg_total = sum(ks_test_gt['is_negative'] == 1)
neg_tp = ks_test_label_res[ks_test_gt['is_negative'] == 1]['is_negative'].sum()
neg_recall_rate = neg_tp / neg_total
print(f'负例整体召回：{neg_recall_rate}')

# 计算各个负面主题的样本数，召回率，准确率，F1, 以及总体macro-F1
topic_neg_num = {label: 0 for label in NEGATIVE_LABELS}
topic_neg_tp = {label: 0 for label in NEGATIVE_LABELS}
topic_neg_pred_num = {label: 0 for label in NEGATIVE_LABELS}
topic_neg_sets = {label: set() for label in NEGATIVE_LABELS}

for index, row in ks_test_gt.iterrows():
    title = row['title']
    if pd.isna(row['negative_labels']):
        continue
    neg_labels = str(row['negative_labels']).split('|')
    for neg_label in neg_labels:
        topic_neg_num[neg_label] += 1
        topic_neg_sets[neg_label].add(title)

for index, row in ks_test_label_res.iterrows():
    title = row['title']
    if pd.isna(row['negative_labels']):
        continue
    neg_labels = str(row['negative_labels']).split('|')
    for neg_label in neg_labels:
        topic_neg_pred_num[neg_label] += 1
        if title in topic_neg_sets[neg_label]:
            topic_neg_tp[neg_label] += 1

macro_f1 = 0
for label in NEGATIVE_LABELS:
    topic_neg_recall = topic_neg_tp[label] / topic_neg_num[label]
    topic_neg_precision = topic_neg_tp[label] / topic_neg_pred_num[label] if topic_neg_pred_num[label] > 0 else 1
    topic_neg_f1 = 2 * topic_neg_precision * topic_neg_recall / (topic_neg_precision + topic_neg_recall)
    macro_f1 += topic_neg_f1
    print(f'{label} 负面样本数：{topic_neg_num[label]}, 召回率：{topic_neg_recall}, 准确率：{topic_neg_precision}, F1：{topic_neg_f1}')
macro_f1 /= len(NEGATIVE_LABELS)
print(f'总体macro-F1：{macro_f1}')
```
